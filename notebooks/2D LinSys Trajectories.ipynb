```python
# A = np.array([[0.5, 0],[0, 0.3]])
# A = np.array([[1.05, 0],[0, 0.5]])
# A = np.array([[1.1, 0],[0, 1.05]])

A = np.array([[0.9, -0.5],[0, 0.9]])

# A = np.array([[0.5, -0.4],[0.4, 0.5]])
# A = np.array([[0.7, -0.8],[0.8, 0.7]])


# A = np.array([[0.5, 1],[0, 0.5]])
# A = np.array([[1.05, 1],[0, 1.05]])
```

```python
x0_list = [np.array([0.1, 0.1]), np.array([0.1, -0.1]), np.array([-0.1, 0.1]), np.array([-0.1, -0.1])]
```

```python
xs_list = []
for x0 in x0_list:
    xs = [x0]
    for _ in range(100):
        xs.append(A @ xs[-1])
    xs = np.array(xs)
    xs_list.append(xs)
```

```python
colors = iter(plt.cm.rainbow(np.linspace(0, 1, len(x0_list))))

plt.figure()
for xs, c in zip(xs_list, colors):
    plt.plot(xs[:,0], xs[:,1], '.', color=c)
    plt.plot(xs[:,0], xs[:,1], alpha=0.5, color=c)
plt.grid()
# plt.axis('equal');
```
