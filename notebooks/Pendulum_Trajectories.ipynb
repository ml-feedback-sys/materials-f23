```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
g = 9.8
l = 2
d = 1
h = 0.1

def F(x):
    theta, omega = x
    return np.array([theta+h*omega, omega+h*(g/l*np.sin(theta)-d*omega)])
```

```python
x0_list = [np.array([0.1, 0.1]), np.array([2, -3]),  np.array([2*np.pi-0.1, -0.1]), np.array([2*np.pi-2, 3]),  ]
```

```python
xs_list = []
for x0 in x0_list:
    xs = [x0]
    for _ in range(150):
        xs.append(F(xs[-1]))
    xs = np.array(xs)
    xs_list.append(xs)
```

```python
colors = iter(plt.cm.rainbow(np.linspace(0, 0.5, len(x0_list))))

plt.figure(figsize=[10,5])
for xs, c in zip(xs_list, colors):
    plt.plot(xs[:,0], xs[:,1], '.', color=c)
    plt.plot(xs[:,0], xs[:,1], alpha=0.5, color=c)
plt.grid()
```
